### Introduction to Questions 5-7

```python
lahman = pd.read_csv("Batting.csv")
```

```python
lahman.head()
```

The Batting data frame contains the offensive statistics for all baseball players over several seasons.  Filter this data frame to define top as the top 10 home run (HR) hitters in 2016:

```python
lahman["yearID"].unique()
```

```python
lahman2016 = lahman[lahman["yearID"] == 2016]
```

```python
lahman2016.head()
```

```python
lahman2016.info()
```

```python
top_names = lahman2016.sort_values(by="HR", ascending=False).head(10)
top_names
```

```python
salary = pd.read_csv("Salaries.csv")
```

```python
salary
```

```python
salary2016 = salary[salary["yearID"] == 2016]
```

```python
salary2016
```

```python
top_salary = pd.merge(left=top_names, right=salary2016, how="inner", on="playerID")
```

```python
top_salary
```

```python
awardplayers = pd.read_csv("AwardsPlayers.csv")
```

```python
awardplayers.head()
```

```python
awardplayers2016 = awardplayers[awardplayers["yearID"] == 2016]
```

```python
awardplayers2016
```

```python
final = pd.merge(left=top_salary, right=awardplayers2016, how="inner", on="playerID")
```

```python
final
```
